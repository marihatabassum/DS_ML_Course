##Cleaning Data!

```python
#filling nan value of Area column
survey_df["Area"].fillna("Rural", inplace=True)

```

```python
#to change to columns numeric!!
survey_df['InternetCost'] = pd.to_numeric(survey_df['InternetCost'], errors='coerce')
survey_df['FamilyIncome'] = pd.to_numeric(survey_df['FamilyIncome'], errors='coerce')
survey_df
```

```python
#filling nan value of InternetCost Column!!
survey_df["InternetCost"].median()
```

```python
survey_df["InternetCost"].fillna(survey_df["InternetCost"].median(), inplace=True)
survey_df
```

```python
#filling the garbage value of InternetCost column!!
survey_df.loc[35, "InternetCost"]=500
survey_df.loc[47, "InternetCost"]=500
```

```python
#filling the garbage value in FamilyIncome column!
survey_df.loc[18, "FamilyIncome"]=50000
```

```python
#now filling the nan value of FamilyIncome after clearing agrbage value! (median will not for the garbage value!
#if it has dtype in object instead it should be integer! we have to change it to numeric dtype then median will work!)
#or you can change the columns into numeric before proceeding!!
survey_df["FamilyIncome"].fillna(survey_df["FamilyIncome"].median(), inplace=True)
```

## Data Transformation!

```python
#using label encoding to process categorical values to other different numeric values!
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder() 
cleaned_df['Gender']= le.fit_transform(cleaned_df['Gender'])
```

```python
cleaned_df
```

```python
#using mapping function in Area column!!
mapping = {'Rural' :1, 'Urban' : 2, "Semi-urban" : 3}
cleaned_df['Area'] = cleaned_df['Area'].map(mapping)
```

```python
cleaned_df
```

```python
#One Hot Encoding for nominal data for Device column!!
one_hot_encoding_device = pd.get_dummies(cleaned_df, columns=['Device'])
one_hot_encoding_device
```

##Normalizing the data!!

```python
#doing standard Scaling!
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
 
standard_scaled_df = pd.DataFrame(std.fit_transform(one_hot_encoding_device), columns=one_hot_encoding_device.columns)
standard_scaled_df
```

```python
#min-max normalizer!!
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(one_hot_encoding_device)
min_max_scaled = pd.DataFrame(scaler.transform(one_hot_encoding_device),columns=one_hot_encoding_device.columns)
min_max_scaled
```

```python
#Robust scaling!!! using quartile range!!
from sklearn.preprocessing import RobustScaler
robust = RobustScaler()
robust_scaled_df = robust.fit_transform(one_hot_encoding_device)
robust_scaled_df = pd.DataFrame(robust_scaled_df, columns=one_hot_encoding_device.columns)
robust_scaled_df
```

```python
#practicing transpose!!
transposed_df=one_hot_encoding_device.transpose()
transposed_df
```
