##### Interval_Estimation: T distribution

check tgis ouT: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html

```python
# method: interval(alpha, df, loc=0, scale=1)
st.t.interval(alpha=0.95, df=len(sam_rural.age)-1, loc=sam_rural.age.mean(), scale=st.sem(sam_rural.age))
```
